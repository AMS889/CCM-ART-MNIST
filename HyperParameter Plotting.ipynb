```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("results/pytorch_results.csv")
```

```python
df_batch = df[(df['dropout_rate']==0.2)&(df['n_filters']==16)&(df['fc_units']==100)&(df['epoch']==15)]
```

```python
plt.figure(figsize=(9,6))
plt.plot('batch_size', 'accuracy', label='1 Layer', data=df_batch[df_batch['layers']==1])
plt.plot('batch_size', 'accuracy', label='2 Layers', data=df_batch[df_batch['layers']==2])
plt.plot('batch_size', 'accuracy', label='3 Layers', data=df_batch[df_batch['layers']==3])
plt.plot('batch_size', 'accuracy', label='4 Layers', data=df_batch[df_batch['layers']==4])
plt.ylabel('Accuracy')
plt.xlabel('Batch Size')
plt.title('Accuracy Versus Batch Size by Model Structure')
plt.legend()

plt.savefig('results/batch_size_tuning.png', dpi = 300)
```

```python
df_drop = df[(df['batch_size']==100)&(df['n_filters']==16)&(df['fc_units']==100)&(df['epoch']==15)]
```

```python
plt.figure(figsize=(9,6))
plt.plot('dropout_rate', 'accuracy', label='1 Layer', data=df_drop[df_drop['layers']==1])
plt.plot('dropout_rate', 'accuracy', label='2 Layers', data=df_drop[df_drop['layers']==2])
plt.plot('dropout_rate', 'accuracy', label='3 Layers', data=df_drop[df_drop['layers']==3])
plt.plot('dropout_rate', 'accuracy', label='4 Layers', data=df_drop[df_drop['layers']==4])
plt.ylabel('Accuracy')
plt.xlabel('Dropout Rate')
plt.title('Accuracy Versus Dropout Rate by Model Structure')
plt.legend()

plt.savefig('results/dropout_tuning.png', dpi = 300)
```

```python
plt.figure(figsize=(9,6))
plt.plot('n_filters', 'accuracy', label='1 Layer', data=df_nfilt[df_nfilt['layers']==1])
plt.plot('n_filters', 'accuracy', label='2 Layers', data=df_nfilt[df_nfilt['layers']==2])
plt.plot('n_filters', 'accuracy', label='3 Layers', data=df_nfilt[df_nfilt['layers']==3])
plt.plot('n_filters', 'accuracy', label='4 Layers', data=df_nfilt[df_nfilt['layers']==4])
plt.ylabel('Accuracy')
plt.xlabel('Number of Filters')
plt.title('Accuracy Versus Number of Filters by Model Structure')
plt.legend()

plt.savefig('results/nfilt_tuning.png', dpi = 300)
```
